```python

import pandas as pd

# Path to raw data
seizure_xlsx = "/Users/mariajosereyesramirez/Documents/CAPP/2ND YEAR/Autumn quarter 2025/Data Viz/interactive_project/interactive_code/raw_data/7.1_Drug_seizures_2019-2023 (1).xlsx"

# 2. Reading Excel File
xls = pd.ExcelFile(seizure_xlsx, engine="openpyxl")
df = pd.read_excel(xls, sheet_name="Seizures")

# 3. Cheking col names
header = df.iloc[0]
df = df.iloc[1:].copy()
df.columns = header

# 4.Filtering only for cocaine
df_cocaine = df[df["DrugGroup"] == "Cocaine-type"].copy()

# 5. Grouping by year and country
agg = (
    df_cocaine
    .groupby(["Reference year", "Country"], as_index=False)["Kilograms"]
    .sum()
)

# 6.Renaming cols
# Here country is region
agg = agg.rename(
    columns={
        "Reference year": "year",
        "Country": "region",  
        "Kilograms": "kg",
    }
)

# 7. Coverting to csv file

agg.to_csv(
    "/Users/mariajosereyesramirez/Documents/CAPP/2ND YEAR/Autumn quarter 2025/Data Viz/interactive_project/interactive_code/data/seizures_cocaine_agg.csv",
    index=False
)

print(agg.head())
```

```python

prev_xlsx = "/Users/mariajosereyesramirez/Documents/CAPP/2ND YEAR/Autumn quarter 2025/Data Viz/interactive_project/interactive_code/raw_data/1.2_Prevalence_of_drug_use_in_the_general_population_national_data (1).xlsx"

# Sheet for cocaine
xls_prev = pd.ExcelFile(prev_xlsx)
coc = pd.read_excel(xls_prev, sheet_name="Cocaine")

# cheking first row for col names
header = coc.iloc[1]
coc2 = coc.iloc[2:].copy()
coc2.columns = header

# Deleting NAN
coc2 = coc2.loc[:, coc2.columns.notna()]

# Country/Territory, Year, Best (prevalencia %)
subset = coc2[["Country/Territory", "Year", "Best"]].copy()

# Testing only for 2 countries (COL, USA)
countries = ["United States of America", "Colombia"]
subset = subset[subset["Country/Territory"].isin(countries)].copy()

# Renaming for JS
subset = subset.rename(
    columns={
        "Country/Territory": "country",
        "Year": "year",
        "Best": "prevalence_15_64",
    }
)

# Sorting by country and year
subset = subset.sort_values(["country", "year"])

# Save
subset.to_csv(
    "/Users/mariajosereyesramirez/Documents/CAPP/2ND YEAR/Autumn quarter 2025/Data Viz/interactive_project/interactive_code/data/prevalence_cocaine_general.csv",
    index=False
)

print(subset.head())
```
